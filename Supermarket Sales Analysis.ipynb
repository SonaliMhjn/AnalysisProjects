Importing libraries

```python
import numpy as np
import pandas as pd
```

Reading the data

```python
df = pd.read_csv('C:/Users/Sonali/Documents/Supermarket_SalesSheet.csv')
df.head(10)
```

Cleaning the data

#Dropping NaN values

```python
nan_df=df[df.isna().any(axis=1)]
nan_df.head(15)
```

```python
df= df.dropna()
df
```

Renaming a Column

```python
df.rename(columns={'Total':'TotalSales'},inplace='True')
```

```python
df.head(10)
```

```python
df['Month']= df['Date'].str[3:5]
df.head()
```

In Which month has the highest sale ?

```python
maxdf= df.groupby('Month')['TotalSales'].sum()
maxdf
```

```python
print("The month with the highest sales is July which is ",maxdf.max())
```

Plotting the Bar Graph

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months , maxdf.values )
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(months)
plt.show()
```

Which city has the highest number of sales?

```python
maxdf = df.groupby('City')['TotalSales'].sum()
maxdf
```

```python
print("City Naypyitaw has the highest sale which is ", maxdf.max())
```

```python
import matplotlib.pyplot as plt
cities = [city for city, df2 in df.groupby('City')]
plt.bar(cities, maxdf.values, color='r' )
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(cities)
plt.show()
```

What  time should we display advertisements to maximize likelihood of customer's buying product?

```python

df['Time'] = pd.to_datetime(df['Time'])

# Extract hour from 'Timestamp'
df['Hours'] = df['Time'].dt.hour
df.head()
```

Plotting Line chart

```python
hours=[hour for hour, df1 in df.groupby('Hours')]
plt.plot(hours,df.groupby(['Hours']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

---At around 19pm that is 7pm is the peak time of the supermarket

Which type of products are sold the most?

```python
df2= df.groupby('Product line')['TotalSales'].sum()
df2
```

Plotting the graph

```python
prod = [prod for prod, df2 in df.groupby('Product line')]
plt.bar(prod, df2.values )
plt.xlabel('Type of Products')
plt.ylabel('Total Sales')
plt.xticks(prod,rotation='vertical')
plt.show()
```

---Foods and Beverages line of product sold the most

Why it got sold the most?

```python
prices = df.groupby('Product line')['Unit price'].mean()
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(prod, df2.values,color='r')
ax2.plot(prod,prices,'b-')
ax1.set_xlabel('Type of Products')
ax1.set_ylabel('Total Sales')
ax2.set_ylabel('Unit Prices',color='b')
ax1.set_xticklabels(prod,rotation='vertical')
plt.show()
```

---It got sold most because it's the basic need specially food 
Moreover prices of foods and beverages is almost declining and are at appropriate rate not too cheap not too costly
